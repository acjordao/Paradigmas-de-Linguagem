## **Questões**

Quais os 10 países que possuem as maiores taxas de desemprego em 2019?


```python
ten_largest = df[['country', '2019']].nlargest(10, '2019')
ten_largest.head(10)
```

```python
fig = plt.figure(figsize=(15, 10))
sns.barplot(
    data=ten_largest,
    x=ten_largest['country'],
    y=ten_largest['2019'],
)
```

Os países que apresentam altas taxas de desemprego tem em comum o fato de serem emergentes, o mercado de trabalho desfuncional indica que as politicas de fomento ao emprego, as lesgislações trabalhistas, as expectativas dos agentes contratantes e a situação economica desses países não estão bem ajustados gerando assim um gap entre a massa de trabalho acima de 15 anos e as vagas ofertadas. A dificuldade de países emergentes em operar em taxas de pleno emprego se dá através desses desalinhamentos que geram problemas no mercado de trabalho.

Quais os 10 países com menores taxas de desemprego em 2019?

```python
ten_smallest = df[['country', '2019']].nsmallest(10, '2019')
ten_smallest.head(10)
```

```python
fig = plt.figure(figsize=(15, 10))
sns.barplot(
    data=ten_smallest,
    x=ten_smallest['country'],
    y=ten_smallest['2019'],
)
```

A recorrencia de países/regiões administrativas asiáticos entre as menores taxas de desemprego pode ser interpretada por um melhor abordagem de politicas de emprego. Mesmo países com menor dessempenho econômico podem apresentar baixas taxas de desemprego é a criação de vagas como principal objetivo das politicas governamentais, sendo por meio de criação de vagas por instituições públicas ou subsidiando empresas.

Qual os 5 países que tiveram as maiores variações de desemprego entre 2018 e 2019?

```python
df ['var'] = df['2019']-df['2018'] 
df
```

```python
five_var_largest = df[['country', 'var']].nlargest(5, 'var')
five_var_largest.head(5)
```

```python
fig = plt.figure(figsize=(15, 10))
sns.barplot(
    data=five_var_largest,
    x=five_var_largest['country'],
    y=five_var_largest['var'],
)
```

Dos cinco países com maior aumento do desemprego entre 2018 e 2019, 4 se encontram na américa latina, a maior voltilidade em número de pessoas empregadas nessa região, pode ser descrita pela depêndencia desse países de suas commodities exportadoras, e como o preço e demanda desses produtos é definido internacionalmente, uma mudança na dinâmica global de compra das commodities principais desses países consegue afetar o desempenho economico e consequentemente as taxas de desemprego desses países.

Quais os 10 países com maiores taxas de desemprego médias entre 2010 e 2019?

```python
df ['mean'] = (df['2019']+df['2018']+df['2017']+df['2016']+df['2015']+df['2014']+df['2013']+df['2012']+df['2011']+df['2010'])/10
df
```

```python
ten_mean_largest = df[['country', 'mean']].nlargest(10, 'mean')
ten_mean_largest.head(10)
```

```python
fig = plt.figure(figsize=(25, 10))
sns.barplot(
    data=ten_mean_largest,
    x=ten_mean_largest['country'],
    y=ten_mean_largest['mean'],
)
```

Ao longo da década entre 2010 e 2019, os países da lista acima tiveram o maior desemprego em média, o que eles tem em comum, são crises de dívida ou instabilidade causada por conflitos ou uma maior instabilidade economica, esse período foi marcante para esses países refletindo em taxas de desemprego maiores, sendo que alguns deles já não estão mais em 2019, entre as maiores taxas de desemprego, demonstrando que o planejamento e execução de medidas que fomentem emprego e melhorem a situação econômica surgem efeito mesmo após grandes eventos.
